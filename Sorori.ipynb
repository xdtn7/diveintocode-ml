```python
import matplotlib.pyplot as plt
import numpy as np
#initial values
no_day = 100
no_rice = [1,]
total_rice = [1,]

for i in range(1, no_day):
    no_rice.append(no_rice[i-1]*2)
    total_rice.append(total_rice[i-1] + no_rice[i])

#print(no_rice[99])

plt.title("Increment of rice")
plt.xlabel("Number of days")
plt.ylabel("Number of grains of rice")
plt.plot(np.arange(1, no_day+1, 1), no_rice, color='red')
plt.plot(np.arange(1, no_day+1, 1),total_rice, color='blue')
plt.legend(['Number of rice grains received', 'Total number of rice grains received'])
plt.grid()
plt.show()


```

```python
def compute_sorori_shinzaemon(day=100):
    
    no_day = day
    list_n_grains = [1,]
    list_total_grains = [1,]

    for i in range(1, no_day):
        list_n_grains.append(no_rice[i-1]*2)
        list_total_grains.append(total_rice[i-1] + no_rice[i])

    plt.title("Increment of grain")
    plt.xlabel("Number of days")
    plt.ylabel("Number of grains of rice")
    plt.plot(np.arange(1, no_day+1, 1), list_n_grains, color='red')
    plt.plot(np.arange(1, no_day+1, 1), list_total_grains, color='blue')
    plt.legend(['Number of rice grains received', 'Total number of rice grains received'])
    plt.grid()
    plt.show()

    return list_n_grains, list_total_grains
list_n_grains, list_total_grains = compute_sorori_shinzaemon(day=10)
print(list_n_grains)
```
