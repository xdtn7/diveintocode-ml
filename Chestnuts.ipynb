# Assignment 3: The chestnut bun problem

## Support libraries and functions

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def print_answer(msg, indent=1, width=None, title=None):
    msg = " ANSWER: " + msg + "  "
    lines = msg.split('\n')
    space = " " * indent
    if not width:
        width = max(map(len, lines))
    box = f'╔{"═" * (width + indent * 2)}╗\n'
    if title:
        box += f'║{space}{title:<{width}}{space}║\n'  # title
        box += f'║{space}{"-" * len(title):<{width}}{space}║\n'  # underscore
    box += ''.join([f'║{space}{line:<{width}}{space}║\n' for line in lines])
    box += f'╚{"═" * (width + indent * 2)}╝'  # lower_border
    print("\n"+box)
```

## [Problem:] If you sprinkle a single chestnut bun with vine, how long will it take to cover the solar system with chestnut buns? Create a program to calculate this.

```python
vol_chestnut = math.pi*(3**2) * 2
dia_solarsys = 287460000000
vol_solarsys = (dia_solarsys/2)**3 * (4/3)* math.pi

no_chest = vol_solarsys / vol_chestnut
period = 5 #minute
i = 1
cnt = 0
while i < no_chest:
    i *= 2
    cnt += 1

period = period * cnt

print_answer("It will take {} minutes to cover the solar system with chestnut buns, which is about {} hours".format(period, period//60))

```

## [Problem:] Create a generic program that can be applied to other situations, such as the time it takes for the Tokyo Dome to fill up when you sprinkle a ball with Byvine. Also, please create a notebook that is easy to understand, e.g., display a graph.

```python
def compute_time(vol_to_fill_, vol_use_to_fill_):
    ratio = vol_to_fill_ / vol_use_to_fill_
    period = 5 #minute
    i = 1
    cnt = 0
    list_vol_use_to_fill = [vol_use_to_fill_,]
    while i < ratio:
        i *= 2
        list_vol_use_to_fill.append(list_vol_use_to_fill[-1]*2)

    plt.title("Increment of volume filled")
    plt.xlabel("Minutes")
    plt.ylabel("Volume filled")
    plt.plot( np.arange(0, len(list_vol_use_to_fill), 1), [vol_to_fill_ for i in range(len(list_vol_use_to_fill))], marker='', color='red', linewidth=2, linestyle='dashed')
    plt.plot( np.arange(0, len(list_vol_use_to_fill), 1), list_vol_use_to_fill, marker='.', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=2)
    plt.legend(['Volume of space being filled', 'Volume filled by Byevine'])
    plt.grid()
    plt.show()

    period = period * len(list_vol_use_to_fill)
    print_answer("It will take {} minutes to cover all the volume with Byevine, which is about {} hours".format(period, period//60))
```

## Run and test:

```python
#For the former example

vol_chestnut = math.pi*(3**2) * 2
dia_solarsys = 287460000000
vol_solarsys = (dia_solarsys/2)**3 * (4/3)* math.pi

compute_time(vol_solarsys, vol_chestnut)
```

```python
#For the latter example

vol_TokyoDome = 4.67 * 10000 * 62   
vol_ball = 0.005792                 

compute_time(vol_TokyoDome, vol_ball)
```
