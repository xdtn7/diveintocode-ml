# Mathematics Lesson for Machine Learning assignment 1 Two-dimensional array and gradient problem

## Support libraries and functions

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ans(msg, indent=1, width=None, title=None):
    msg = " ANSWER: " + msg + "  "
    lines = msg.split('\n')
    space = " " * indent
    if not width:
        width = max(map(len, lines))
    box = f'╔{"═" * (width + indent * 2)}╗\n'
    if title:
        box += f'║{space}{title:<{width}}{space}║\n'  # title
        box += f'║{space}{"-" * len(title):<{width}}{space}║\n'  # underscore
    box += ''.join([f'║{space}{line:<{width}}{space}║\n' for line in lines])
    box += f'╚{"═" * (width + indent * 2)}╝'  # lower_border
    print("\n"+box)
```

## [Problem 1] Linear function

```python
x = np.arange(-50, 50.1, 0.1)
y = 1/2 * x + 1
ans("x = {}\n y = {}".format(x, y))
```

## [Problem 2] Array combination

```python
z = np.column_stack((x, y))
ans("z shape = {} \n z = {}".format(z.shape, z))
```

## [Problem 3] Find the gradient

```python
grdn = (y[1:] - y[:-1])/(x[1:] - x[:-1])
ans("Gradient: {}".format(grdn))
```

## [Problem 4] Draw a graph

```python

fig, ax = plt.subplots(2, 1, sharex = True)
ax[0].set_ylabel("y")
ax[0].set_title("Linear function")
ax[0].plot(z[:, 0], z[:, 1])
ax[0].grid()
ax[1].set_xlabel("x")
ax[1].set_ylabel("Gradient")
ax[1].set_title("Gradient")
ax[1].plot(z[: -1, 0], grdn)
ax[1].grid()
plt.show()
```

## [Problem 5] Python functionalization

```python
def compute_gradient(function, x_range):
    array_xy = np.column_stack((x_range, function(x_range)))
    gradient = (function(x_range)[1:] - function(x_range)[:-1])/(x_range[1:] - x_range[:-1])
    return array_xy, gradient

def draw_graphs(num, array_xy, gradient):
  print("Graphs of function {}: \n".format(num))
  fig, ax = plt.subplots(2, 1, sharex = True)
  ax[0].set_ylabel("y")
  ax[0].set_title("Linear function")
  ax[0].plot(array_xy[:, 0], array_xy[:, 1])
  ax[0].grid()
  ax[1].set_xlabel("x")
  ax[1].set_ylabel("Gradient")
  ax[1].set_title("Gradient")
  ax[1].plot(array_xy[: -1, 0], gradient)
  ax[1].grid()
  plt.show()

def function1(arr_x_):
  return  arr_x_**2

def function2(arr_x_):
  return 2*arr_x_**2 + 2**arr_x_

def function3(arr_x_):
  return np.sin(arr_x_**(1/2))

array_xy1, gradient1 = compute_gradient(function1, np.arange(-50, 50.1, 0.1))
array_xy2, gradient2 = compute_gradient(function2, np.arange(-50, 50.1, 0.1))
array_xy3, gradient3 = compute_gradient(function3, np.arange(0, 50.1, 0.1))

draw_graphs(1, array_xy1, gradient1)
draw_graphs(2, array_xy2, gradient2)
draw_graphs(3, array_xy3, gradient3)
```

## [Problem 6] Find the minimum value

```python
def draw_min(num, array_xy, gradient, min):
    print("Graphs of function {}: \n".format(num))
    plt.ylabel("y")
    plt.xlabel("x")
    plt.title("Linear function")
    plt.plot(array_xy[:, 0], array_xy[:, 1])
    plt.axhline(min, color='r', linestyle='-')
    plt.grid()
    plt.show()
```

## $y = x^2 (−50 ≤ x ≤ 50)$

```python
min1 = round(np.min(array_xy1[:, 1]), 4)
ans("Minimum value of y = {} at x = {}".format(min1, np.argmin(array_xy1[:, 1])*0.1 + (-50)))
```

```python
draw_min(1, array_xy1, gradient1, min1)
```

## $y = 2x^2 + 2x (−50 ≤ x ≤ 50)$

```python
min2 = round(np.min(array_xy2[:, 1]), 4)
ans("Minimum value of y = {} at x = {}".format(min2, np.argmin(array_xy2[:, 1])*0.1 + (-50)))
```

```python
draw_min(2, array_xy2, gradient2, min2)
```

## $y = sin(x^\frac{1}{2}) (0 ≤ x ≤ 50)$

```python
min3 = round(np.min(array_xy3[:, 1]), 4)
ans("Minimum value of y = {} at x = {}".format(min3, np.argmin(array_xy3[:, 1])*0.1 + (-50)))
```

```python
draw_min(3, array_xy3, gradient3, min3)
```
