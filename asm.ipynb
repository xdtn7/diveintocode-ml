```python
import time

THICKNESS = 0.00008
n = 43
coef = 2
start = time.time()

folded_thickness = THICKNESS*(coef**n)

mid_time = time.time()

for i in range (1, n):
    coef = coef* 2
folded_thickness = THICKNESS*coef

elapsed_time2 = time.time() - mid_time
elapsed_time1 = mid_time-start
print("time (arithmetic operators) : {}[s]".format(elapsed_time1))
print("time (for statement) : {}[s]".format(elapsed_time2))
```

```python
THICKNESS = 0.00008
n = 43
coef = 2
e = []
folded_thickness = THICKNESS
e.append(folded_thickness)
for i in range (0, n):
    folded_thickness = folded_thickness*coef
    e.append(folded_thickness)

print(len(e))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(e) # Enter the variable name of the list in "List name"
plt.show()

#brief description: The graph increases exponetially
```
